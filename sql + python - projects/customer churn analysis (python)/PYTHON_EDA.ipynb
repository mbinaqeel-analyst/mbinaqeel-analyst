```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Customer Churn.csv")
```

Starting inspection of data

```python
df.info()
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
df.info()
```

Total Charges converted datatype from object to float

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if (value == 1):
        return "yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df.head(50)
```

Converted senior citizen dtype to Object

```python
plt.figure(figsize=(6,4))  # adjust size for clarity
ax = sns.countplot(x="Churn", data=df, palette="pastel")  # add colors
plt.title("Customer Churn Distribution", fontsize=14)
plt.xlabel("Churn", fontsize=12)
plt.ylabel("Count", fontsize=12)
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
ax = sns.countplot(x="gender", hue="Churn", data=df, palette="pastel")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churn by Gender")
plt.show()
```

```python
sns.countplot(x="SeniorCitizen", hue="Churn", data=df, palette="pastel")
plt.title("Churn by Senior Citizen Status")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="tenure", hue="Churn", kde=True, bins=30, palette="pastel")

plt.title("Customer Tenure Distribution by Churn", fontsize=14)
plt.xlabel("Tenure (Months)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Churn", labels=["No", "Yes"])  # custom legend labels
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="Contract", hue="Churn", multiple="dodge", shrink=0.8, palette="pastel")

plt.title("Churn by Contract Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(18, 12))

# 1 - Payment Method
plt.subplot(2, 3, 1)
sns.countplot(data=df, x="PaymentMethod", hue="Churn", palette="Set2")
plt.title("Churn by Payment Method")
plt.xticks(rotation=45)

# 2 - Tech Support
plt.subplot(2, 3, 2)
sns.countplot(data=df, x="TechSupport", hue="Churn", palette="Set2")
plt.title("Churn by Tech Support")

# 3 - Internet Service
plt.subplot(2, 3, 3)
sns.countplot(data=df, x="InternetService", hue="Churn", palette="Set2")
plt.title("Churn by Internet Service")

# 4 - Online Security
plt.subplot(2, 3, 4)
sns.countplot(data=df, x="OnlineSecurity", hue="Churn", palette="Set2")
plt.title("Churn by Online Security")

# 5 - Streaming TV
plt.subplot(2, 3, 5)
sns.countplot(data=df, x="StreamingTV", hue="Churn", palette="Set2")
plt.title("Churn by Streaming TV")

# 6 - Monthly Charges (numeric)
plt.subplot(2, 3, 6)
sns.histplot(data=df, x="MonthlyCharges", hue="Churn", bins=30, kde=True, palette="Set2")
plt.title("Monthly Charges by Churn")

plt.tight_layout()
plt.show()
```
